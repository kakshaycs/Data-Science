```python
# to extract top "length" user,movies those have high frequency 

X=data["user"].value_counts()    # pandas series return
Y=data['movie'].value_counts()   # pandas series return
```

```python
 # extract the index of the series that is either user id or movie id
length=50
users=X.index      
movies=Y.index
users=user[:length]
movies=movies[:length]
```

```python
# this is the Rating matrix that contain column is user and rows is movies
"""    | movie1 | movie2 | movie3 
 user1  |   5    |    4   |    1
 user2  |   3    |        |    5
 user3  |   5    |    3   |    
  
"""
# the R matrix is boolean matrix that denote user rated that movie or not

Rating=[[0 for _ in range(length)] for _ in range(length)]
R=[[0 for _ in range(length)] for _ in range(length)]
```

```python
# update the both matrix Rating and R
for i in range(length):
    for j in range(length):
        a=data[(data['user']==users[i]) & (data['movie']==movies[j])]
        if a.empty!=True:
            Rating[i][j]=list(a['rating'])[0]  
            R[i][j]=1
        clear_output(wait=True)
        print ("percentage = ",i*j/25)
```

```python
DataRating=pd.DataFrame(Rating)
DataR=pd.DataFrame(R)
```
